```python
import pandas as pd
```

```python
df=pd.read_csv('preprocessed_dataset(1).csv')
```

```python
gender_counts = df['Gender'].value_counts()
print(gender_counts)
```

## Gender

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.show()
```

## Age group

```python
age_group_counts = df['Age group'].value_counts().sort_index()

# Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(age_group_counts.index, age_group_counts.values, color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Age Group Distribution')
plt.xticks(rotation=45)
plt.show()

# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(age_group_counts))))
plt.title('Age Group Distribution')
plt.show()
```

## Marital status

```python
# Filter out 'Not disclosed' category
df_filtered = df[df['Marital status'] != 'Not disclosed']

# Set up the figure
plt.figure(figsize=(10, 6))

# Create the countplot
sns.countplot(data=df_filtered, x='Marital status', palette='viridis')

# Customize plot
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.title('Marital Status Distribution')
plt.xticks(rotation=45)

# Show plot
plt.show()
```

## Occupation plot

```python
# Filter out 'Unknown' category
df_filtered = df[df['Occupation'] != 'Unknown']
df_filtered = df[df['Occupation'] != 'Others']

# Set up the figure
plt.figure(figsize=(10, 6))

# Create the countplot
sns.countplot(data=df_filtered, x='Occupation', palette='viridis')

# Customize plot
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Occupation Distribution')
plt.xticks(rotation=45)

# Show plot
plt.show()
```

## State

```python
# Filter out irrelevant categories
states_filtered = df[~df['State'].str.contains('Total')]

# Calculate the counts for states
state_counts = states_filtered['State'].value_counts()

# Pie Chart
plt.figure(figsize=(10, 10))
plt.pie(state_counts, labels=state_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(state_counts))))
plt.title('State Distribution')
plt.show()
```

## Sleeping time

```python
# Filter out 'Unknown' category
filtered_df = df[df['Sleeping time'] != 'Unknown']

# Convert sleeping time to numeric (if necessary)
filtered_df['Sleeping time'] = pd.to_numeric(filtered_df['Sleeping time'], errors='coerce')

# Set up the figure
plt.figure(figsize=(10, 6))

# Create the box plot
sns.boxplot(x=filtered_df['Sleeping time'], color='skyblue')

# Customize plot
plt.xlabel('Sleeping Time (hours)')
plt.title('Box Plot of Sleeping Time')

# Show plot
plt.show()
```

```python
import pandas as pd'''  '''
```

```python
data=pd.read_csv('preprocessed_dataset.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

##  Pie charts for gender distribution within each mental health issue category

```python

# Pie charts for gender distribution within each mental health issue category
for state in data['Depression State'].unique():
    plt.figure(figsize=(6, 6))
    data[data['Depression State'] == state]['Gender'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='viridis')
    plt.title(f'Gender Distribution for {state} Depression')
    plt.ylabel('')
    plt.show()
```

##  Histogram showing the frequency of age groups for each depression state

```python
# Filter out 'prefer not to disclose' from the 'Depression State' column
data_filtered = data[data['Depression State'] != 'prefer not to disclose']

# Convert numerical variables to numeric data types if they are not already
data_filtered['Concentration'] = pd.to_numeric(data_filtered['Concentration'], errors='coerce')

# Histogram showing the frequency of age groups for each depression state
plt.figure(figsize=(10, 6))
sns.histplot(data=data_filtered, x='Age group', hue='Depression State', multiple='stack', shrink=0.8)
plt.title('Age Group Distribution by Depression State')
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```


## Bar plot to show the average concentration for each depression state

```python
# Bar plot to show the average concentration for each depression state
plt.figure(figsize=(10, 6))
sns.barplot(data=data_filtered, x='Depression State', y='Concentration', ci=None)
plt.title('Average Concentration for Each Depression State')
plt.xlabel('Depression State')
plt.ylabel('Average Concentration')
plt.tight_layout()
plt.show()
```

## relationship between marital status and depression states

```python
# Filter out 'Not disclosed' from the 'Marital status' column and 'prefer not to disclose' from the 'Depression State' column
data_filtered = data[(data['Marital status'] != 'Not disclosed') & (data['Depression State'] != 'prefer not to disclose')]

# Convert numerical variables to numeric data types if they are not already
data_filtered['Concentration'] = pd.to_numeric(data_filtered['Concentration'], errors='coerce')

# Grouped bar chart to show the relationship between marital status and depression states
marital_depression_counts = data_filtered.groupby(['Marital status', 'Depression State']).size().unstack(fill_value=0)
plt.figure(figsize=(10, 6))
marital_depression_counts.plot(kind='bar', stacked=True)
plt.title('Marital Status and Depression States')
plt.xlabel('Marital Status')
plt.ylabel('Frequency')
plt.legend(title='Depression State')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Age VS Marital status VS DepressionState Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the font size and figure size
sns.set_context("talk", font_scale=1.2)
plt.figure(figsize=(20, 12))

# Filter out 'prefer not to disclose' from the 'Depression State' column, 'Not disclosed' from the 'Marital status' column, and 'No depression' from the 'Depression State' column
data_filtered = data[(data['Depression State'] != 'prefer not to disclose') &
                     (data['Marital status'] != 'Not disclosed') &
                     (data['Depression State'] != 'No depression')]

# Convert numerical variables to numeric data types if they are not already
data_filtered['Concentration'] = pd.to_numeric(data_filtered['Concentration'], errors='coerce')

# Faceted bar plot to show the relationship between age group, marital status, and depression state
g = sns.FacetGrid(data_filtered, col="Marital status", hue="Depression State", col_wrap=3, height=5, aspect=2, palette='Set2')
g.map(sns.histplot, "Age group", multiple="stack", shrink=0.8, bins=10)
g.add_legend()
g.set_titles("{col_name}")
g.set_axis_labels("Age Group", "Frequency")
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Age Group Distribution by Marital Status and Depression State', fontsize=20)
plt.tight_layout()

# Save the figure
# g.savefig('age_marital_depression_distribution.png')

# Show the plot
plt.show()
```

## relationship between occupation types and depression states

```python
# Grouped bar chart to show the relationship between occupation types and depression states
occupation_depression_counts = data_filtered.groupby(['Occupation', 'Depression State']).size().unstack(fill_value=0)
occupation_depression_counts.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Occupation-wise Distribution of Depression States')
plt.xlabel('Occupation')
plt.ylabel('Frequency')
plt.legend(title='Depression State')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## show the effect of meditation and exercises on depression states

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out 'Unknown/Uncertain' from the 'Meditation and Exercises' column and 'prefer not to disclose' from the 'Depression State' column
data_filtered = data[(data['Meditation and Exercises'] != 'Unknown/Uncertain') & (data['Depression State'] != 'prefer not to disclose')]

# Grouped bar chart to show the effect of meditation and exercises on depression states
exercise_depression_counts = data_filtered.groupby(['Meditation and Exercises', 'Depression State']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 6))
exercise_depression_counts.plot(kind='bar', stacked=True)
plt.title('Impact of Meditation and Exercises on Depression States')
plt.xlabel('Meditation and Exercises')
plt.ylabel('Frequency')
plt.legend(title='Depression State')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Scatter plot to show the relationship between physical activities and mental health parameters like concentration and hopelessness
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data_filtered, x='Concentration', y='Hopelessness', hue='Meditation and Exercises', style='Depression State')
plt.title('Relationship Between Physical Activities and Mental Health Parameters')
plt.xlabel('Concentration')
plt.ylabel('Hopelessness')
plt.tight_layout()
plt.show()
```

```python
# Filter out 'Unknown/Uncertain' from the 'Meditation and Exercises' column and 'prefer not to disclose' from the 'Depression State' column
data_filtered = data[(data['Meditation and Exercises'] != 'Unknown/Uncertain') & (data['Depression State'] != 'prefer not to disclose')]

# Grouped data for the line plot
exercise_depression_counts = data_filtered.groupby(['Meditation and Exercises', 'Depression State']).size().unstack(fill_value=0)

# Plotting the line plot
plt.figure(figsize=(12, 6))
for column in exercise_depression_counts.columns:
    plt.plot(exercise_depression_counts.index, exercise_depression_counts[column], marker='o', label=column)

plt.title('Impact of Meditation and Exercises on Depression States')
plt.xlabel('Meditation and Exercises')
plt.ylabel('Frequency')
plt.legend(title='Depression State')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Relationship Between Sleeping Time and Depression States

```python
# Exclude 'Prefer not to disclose' from the dataset
filtered_data_depression = data[(data['Depression State'] != 'Prefer not disclose') & (data['Sleeping time'] != 'not disclose')]

plt.figure(figsize=(10, 6))
sns.histplot(data=filtered_data_depression, x='Sleeping time', hue='Depression State', multiple='stack', shrink=0.8)
plt.title('Sleeping Time Distribution by Depression State')
plt.xlabel('Sleeping Time')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Percentage of People Taking Treatments for Mental Health Issues

```python
# Exclude 'Prefer not to disclose' from 'taking any treatment?' column
filtered_treatment_data = data[data['Phys_health_interview'] != 'Prefer not to disclose']

# Pie chart
treatment_counts = filtered_treatment_data['taking any treatment?'].value_counts()

plt.figure(figsize=(6, 6))
treatment_counts.plot.pie(autopct='%1.1f%%', startangle=90, cmap='viridis')
plt.title('Percentage of People Taking Treatments for Mental Health Issues')
plt.ylabel('')
plt.show()
```

##  Relationship Between Physical Health Interview Outcomes and Depression States

```python
import seaborn as sns

# Exclude 'Prefer not to disclose' from 'Phys_health_interview' and 'Depression State'
filtered_health_depression_data = data[
    (data['Phys_health_interview'] != 'prefer not to disclose') &
    (data['Depression State'] != 'prefer not to disclose')
]

# Group by 'Phys_health_interview' and 'Depression State'
physical_health_depression_counts = filtered_health_depression_data.groupby(['Phys_health_interview', 'Depression State']).size().unstack(fill_value=0)

# Bar chart
plt.figure(figsize=(10, 6))
physical_health_depression_counts.plot(kind='bar', stacked=True)
plt.title('Physical Health Interview Outcomes and Depression States')
plt.xlabel('Physical Health Interview')
plt.ylabel('Frequency')
plt.legend(title='Depression State')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Convert categorical data to numerical data for correlation analysis
df_encoded = data.copy()
for col in df_encoded.columns:
    if df_encoded[col].dtype == 'object':
        df_encoded[col] = df_encoded[col].astype('category').cat.codes

# Heatmap of Correlation Matrix
plt.figure(figsize=(14, 10))
sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
